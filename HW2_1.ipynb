##Function 1 and Point 1 (minimum)

```
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-white')

F1 = plt.axes(projection='3d')
def f(x, y):
    return x+y

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
xdata = 0
ydata = 0
zdata = 0

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z, 40, cmap='RdGy')
plt.colorbar()
F1.plot3D(X.flatten(), Y.flatten(), Z.flatten(), label='F1')
F1.scatter(xdata, ydata, zdata, label='P1', color='red')
F1.legend()
```

##Function 2 and Point 2 (maximum)

```
F2 = plt.axes(projection='3d')
def g(x, y):
    return y ** 2 + x

x = np.linspace(0, 15, 50)
y = np.linspace(0, 15, 40)
xdata = 15
ydata = 15
zdata = 240

X, Y = np.meshgrid(x, y)
Z = g(X, Y)
plt.contour(X, Y, Z, 40, cmap='RdGy')
plt.colorbar()
F2.plot3D(X.flatten(), Y.flatten(), Z.flatten(), label='F2')
F2.scatter(xdata, ydata, zdata, label='P2', color='red')
F2.legend()
```

##Function 3 and Point 3 (neither a max nor a min, but gradient is 0)

```
F3 = plt.axes(projection='3d')


def h(x, y):
    return (x ** 2) - (y ** 2)

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 40)
xdata = 0
ydata = 0
zdata = 0

X, Y = np.meshgrid(x, y)
Z = h(X, Y)
plt.contour(X, Y, Z, 40, cmap='RdGy')
plt.colorbar()
F3.view_init(0, 20)
F3.plot3D(X.flatten(), Y.flatten(), Z.flatten(), label='F3')
F3.scatter(xdata, ydata, zdata, label='P3', color='red')
F3.legend()
F3
```
